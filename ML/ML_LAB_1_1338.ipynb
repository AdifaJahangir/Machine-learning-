```python
#loading data and showing features
from sklearn.datasets import load_iris
iris=load_iris()
print(iris.keys())
print(iris["target_names"])
n_samples,n_features=iris.data.shape
print("Number of samples:",n_samples)
print("Number of features:",n_features)
print('feature Name:',iris.feature_names)
print("Target Name:",iris.target_names)
print("Dimension of Input",iris.data.shape)
print("Dimension of Output:",iris.target.shape)
print("First five rows of Iutput:",iris.data[1:5])
print("Target Data:",iris.target)
```

```python
#plotting graphs ko plot kr rhy
#histogram of the features
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
x_index=3
colors=['blue','red','green']
for label,color in zip(range(len(iris.target_names)),colors):
 ax.hist(iris.data[iris.target==label,x_index],label=iris.target_names[label],color=color)
 ax.set_xlabel(iris.feature_names[x_index])
 ax.legend(loc='upper right')
fig.show()
```

```python
#Scatterplot with two features
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

# Load the Iris dataset
iris = load_iris()

# Define feature indices for plotting
x_index = 3  # Petal width
y_index = 0  # Sepal length
colors = ['blue', 'green', 'red']

# Create a figure and axis
fig, ax = plt.subplots()

# Scatter plot for each class
for label, color in zip(range(len(iris.target_names)), colors):
    ax.scatter(
        iris.data[iris.target == label, x_index],
        iris.data[iris.target == label, y_index],
        label=iris.target_names[label],
        c=color
    )

# Set labels and legend
ax.set_xlabel(iris.feature_names[x_index])
ax.set_ylabel(iris.feature_names[y_index])
ax.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
#Scatter plot Total

import matplotlib.pyplot as plt
n=len(iris.feature_names)
fig,ax=plt.subplots(n,n,figsize=(15,15))
colors=['blue','red','green']
for x in range (n):
  for y in range (n):
    xname=iris.feature_names[x]
    yname=iris.feature_names[y]
    for color_ind in range(len(iris.target_names)):
     ax[x,y].scatter(iris.data[iris.target==color_ind,x],iris.data[iris.target==color_ind,y],label=iris.target_names[color_ind],c=colors[color_ind])
     ax[x,y].set_xlabel(xname)
     ax[x,y].set_ylabel(yname)
     ax[x,y].legend(loc='upper left')
  plt.show()
```

```python
#Activity 3
#Scatter Plot
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['target'] = wine.target  # Adding target column
X = df.drop(columns=['target'])
y = df['target']
import matplotlib.pyplot as plt
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, cmap='viridis')
plt.xlabel(wine.feature_names[0])
plt.ylabel(wine.feature_names[1])
plt.title('Scatter Plot')
plt.show()
```
